```python
import numpy as np

fpath = 'annotated/'
name = filename[0]
ind = 1

def getFilePath(name, ind):
    orig = fpath + name + str(ind) + '.txtoriginal.txt'
    anno = fpath + name + str(ind) + '.txt'
    return {'originalFilePath': orig, 'annotedFilePath' : anno}


cct = CCT(**getFilePath(name, 3))
cct.execute()
cct.getAnnotedEntities('RTag')
cct.coporAnnotation('RTag')
cct.getAnnotedEntities('ETag')
cct.coporAnnotation('ETag')
cct.dfFormat
df = cct.dfFormat
df.loc[len(df)] = np.NaN
df
```

```python
for name in filename:
    for ind in range(300):
        print(getFilePath(name, ind))
```

```python
d = getFilePath(name, ind+1)
for i in d.values():
    os.path.isfile(i)
```

```python
import time
btime = time.clock()
timeConsmpt = []
CCTList = []
Missing = []
for name in filename:
    #name = filename[2]
    for ind in range(300):
        bbtime = time.clock()
        ## To check whether a file exists.
        d = getFilePath(name, ind+1)
        flg = True
        for i in d.values():
            if not os.path.isfile(i):
                flg = False
                Missing.append(i)
        if not flg:
            continue
        print(ind+1)
        cct = CCT(**getFilePath(name, ind+1))
        cct.execute()
        CCTList.append(cct)
        eetime = time.clock()
        timeConsmpt.append([name+str(ind+1),eetime - bbtime])
        
etime = time.clock()
```

```python
print(etime - btime)
```

```python
print('Time Consumption: ', (etime - btime)/60,'mins')
```

```python
Missing
```

```python
%matplotlib inline
import pandas as pd

time = pd.Series([i[-1] for i in timeConsmpt])
```

```python
time.plot(figsize=(12,3), xticks = [i*100 for i in list(range(13))])
```

### Generate a Dict

```python
filename = ['01-一般项目/一般项目-',
            '02-病史特点/病史特点-',
            '04-诊疗经过/诊疗经过-',
            '05-出院情况/出院情况-']

types = [item.split('/')[-1].replace('-', '') for item in filename]

cctDict = {}
for t in types:
    cctDict[t] = []
    
for cct in CCTList:
    for t in types:
        if t in cct.originalFilePath:
            cctDict[t].append(cct)
            break
```

```python
len(cctDict[types[0]])
```

### Serialize CCT to Pickle

```python
import pickle
import os


from datetime import datetime
now = datetime.now()
date = now.date()
print(date)


#os.mkdir('pkldata/'+ str(date))

with open('pkldata/'+ str(date) + '/CCT_List'+'.p', 'wb') as handle:
    pickle.dump(CCTList, handle)
    
with open('pkldata/'+ str(date) + '/CCT_Dict'+'.p', 'wb') as handle:
    pickle.dump(cctDict, handle)

with open('pkldata/'+ str(date) + '/CCT_Log' +'.p', 'wb') as handle:
    pickle.dump([time, timeConsmpt,Missing], handle)
```

### Tags On



```python
import time
btime = time.clock()
timeConsmpt = []
CCTList = []
Missing = []
for name in filename:
    #name = filename[2]
    for ind in range(300):
        bbtime = time.clock()
        ## To check whether a file exists.
        d = getFilePath(name, ind+1)
        flg = True
        for i in d.values():
            if not os.path.isfile(i):
                flg = False
                Missing.append(i)
        if not flg:
            continue
        print(ind+1)
        cct = CCT(**getFilePath(name, ind+1))
        cct.execute()
        cct.getAnnotedEntities('RTag')
        cct.coporAnnotation('RTag')
        cct.getAnnotedEntities('ETag')
        cct.coporAnnotation('ETag')
        CCTList.append(cct)
        eetime = time.clock()
        timeConsmpt.append([name+str(ind+1),eetime - bbtime])
        
etime = time.clock()
```

```python
print(etime - btime)
print('Time Consumption: ', (etime - btime)/60,'mins')
```

```python
%matplotlib inline
import pandas as pd

time = pd.Series([i[-1] for i in timeConsmpt])
time.plot(figsize=(12,3), xticks = [i*100 for i in list(range(13))])
```

```python
filename = ['01-一般项目/一般项目-',
            '02-病史特点/病史特点-',
            '04-诊疗经过/诊疗经过-',
            '05-出院情况/出院情况-']

types = [item.split('/')[-1].replace('-', '') for item in filename]

cctDict = {}
for t in types:
    cctDict[t] = []
    
for cct in CCTList:
    for t in types:
        if t in cct.originalFilePath:
            cctDict[t].append(cct)
            break
```

```python
import pickle
import os


from datetime import datetime
now = datetime.now()
date = now.date()
print(date)


try:
    os.mkdir('pkldata/'+ str(date))
except:
    pass

with open('pkldata/'+ str(date) + '/R_CCT_List'+'.p', 'wb') as handle:
    pickle.dump(CCTList, handle)
    
with open('pkldata/'+ str(date) + '/R_CCT_Dict'+'.p', 'wb') as handle:
    pickle.dump(cctDict, handle)

with open('pkldata/'+ str(date) + '/R_CCT_Log' +'.p', 'wb') as handle:
    pickle.dump([time, timeConsmpt,Missing], handle)
```
